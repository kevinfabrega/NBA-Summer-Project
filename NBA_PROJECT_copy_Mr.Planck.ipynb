```python
Team=df['Team_per_game'].unique()
Team
eliminar = ['2TM', '3TM', '4TM', '5TM']
Team_f= Team[~np.isin(Team, eliminar) & ~pd.isnull(Team)]
Team_f
```

```python
###################################################################
equivalencias = {
    'NOH': 'NOP',
    'NOK': 'NOP',
    'CHA': 'CHO',
    'CHH': 'CHO',
    'SEA': 'OKC',
    'NJN': 'BRK',
    'VAN': 'MEM',
}
arr_unificado = np.array([equivalencias.get(equipo, equipo) for equipo in Team_f])
equipos_unicos = sorted(np.unique(arr_unificado))

# Crear diccionario de √≠ndices para codificaci√≥n one-hot
equipo_a_idx = {equipo: idx for idx, equipo in enumerate(equipos_unicos)}

# Codificar cada equipo como vector one-hot. Aquesta funci√≥ rep les abreviatures DIRECTAMENT i les transforma en la llista dessitjada
def one_hot(equipo):
    vec = [0] * len(equipos_unicos)
    idx = equipo_a_idx[equivalencias.get(equipo, equipo)]
    vec[idx] = 1
    return vec
###################################################################

# Lista de vectores one-hot
vectores_one_hot = [one_hot(equipo) for equipo in Team_f]

# Ejemplo: mostrar el primero
for i in range(len(Team_f)):
  print(f"'{Team_f[i]}' ‚Üí {vectores_one_hot[i]}")
```

```python
LeBron['team'] = LeBron.iloc[:,1].apply(one_hot)
LeBron = LeBron.drop(columns='Team_per_game')
LeBron
```

```python
################################################################### El mateix dels equips per√≤ amb les posicions
arr_unificado = np.array([equivalencias.get(equipo, equipo) for equipo in Team_f])
equipos_unicos = sorted(np.unique(arr_unificado))
pos_possibles=['PG','SG','SF','PF','C']
# Crear diccionario de √≠ndices para codificaci√≥n one-hot
pos_a_idx = {pos: idx for idx, pos in enumerate(pos_possibles)}

def one_hotp(pos):
    vec = [0] * len(pos_possibles)
    idx1 = pos_a_idx[pos]
    vec[idx1] = 1
    return vec
###################################################################
```

```python
LeBron['pos'] = LeBron.iloc[:,1].apply(one_hotp)
LeBron = LeBron.drop(columns='Pos_per_game')
LeBron
```
